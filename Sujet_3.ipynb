## Sujet N¬∞3. ü§ñ Optimisation de Dispatch de T√¢ches (Type Uber)

#### √âtape 1 : Mod√©lisation des Donn√©es (List, Dict, Numpy)

```python
import numpy as np

vehicules = [
    {'id': 1, 'position': np.array([0, 0]), 'disponible': True},
    {'id': 2, 'position': np.array([5, 3]), 'disponible': True},
    {'id': 3, 'position': np.array([2, 8]), 'disponible': True},
    {'id': 4, 'position': np.array([7, 1]), 'disponible': True},
    {'id': 5, 'position': np.array([1, 4]), 'disponible': True}
]

clients_en_attente = [
    {'id_client': 101, 'position': np.array([3, 2])},
    {'id_client': 102, 'position': np.array([6, 5])},
    {'id_client': 103, 'position': np.array([1, 1])},
    {'id_client': 104, 'position': np.array([8, 7])},
    {'id_client': 105, 'position': np.array([4, 3])}
]
```

#### √âtape 2 : Fonction Utilitaire (Fonction, Numpy)

```python
def calculer_distance(pos1, pos2, type_distance="euclidienne"):
    if type_distance == "euclidienne":
        return np.linalg.norm(pos1 - pos2)
    elif type_distance == "manhattan":
        return np.sum(np.abs(pos1 - pos2))
    else:
        raise ValueError("Type de distance non reconnu.")
```

#### √âtape 3 : Boucle de Dispatch (Fonctions Avanc√©es, for, min)

```python
assignations = []

for client in clients_en_attente:
    # 1Ô∏è Filtrage des v√©hicules disponibles
    vehicules_dispos = list(filter(lambda v: v["disponible"], vehicules))

    # 2Ô∏è Si aucun v√©hicule n‚Äôest disponible, on arr√™te
    if not vehicules_dispos:
        print("Plus de v√©hicules disponibles !")
        break

    # 3Ô∏è Calcul des distances client -> v√©hicules
    distances = list(map(lambda v: calculer_distance(client["position"], v["position"]), vehicules_dispos))

    # 4Ô∏è Association distances <-> v√©hicules
    distances_zip = list(zip(distances, vehicules_dispos))

    # 5Ô∏è Trouver le v√©hicule le plus proche
    distance_min, vehicule_elu = min(distances_zip, key=lambda x: x[0])

    # 6Ô∏è Mise √† jour de l'√©tat et enregistrement
    print(f"Client {client['id_client']} -> V√©hicule {vehicule_elu['id']} (distance = {distance_min:.2f})")
    assignations.append((client["id_client"], vehicule_elu["id"]))
    vehicule_elu["disponible"] = False

```

#### √âtape 4 : Affichage des R√©sultats (print)

```python
print(" Liste finale des assignations :")
for c, v in assignations:
    print(f" - Client {c} -> V√©hicule {v}")

print(" √âtat final des v√©hicules :")
for v in vehicules:
    print(f"V√©hicule {v['id']} | Position {v['position']} | Disponible : {v['disponible']}")
```
